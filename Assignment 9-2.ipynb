```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
# Importing the data
movies_data = pd.read_csv('/home/suraj/Downloads/Assignments/Assignment-9/my_movies.csv')
movies_data
```

```python
movies_data1 = movies_data.iloc[:,5:]
movies_data1
```

```python
movies_data1.info()
```

```python
# Model building
# With 10% Support and 70% confidence

frequent_item_sets = apriori(df =movies_data1 ,min_support=0.1, use_colnames=True) 
frequent_item_sets
```

```python
rules = association_rules(df = frequent_item_sets ,metric='confidence',min_threshold=0.7)
rules
```

```python
# visualization of obtained rule
plt.scatter(rules['support'],rules['confidence'],c=rules['lift'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
